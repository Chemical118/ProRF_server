```julia
include("load_save.jl")
dataset_name = String(split(split(@__FILE__, '\\')[end], '.')[1])
R, X, Y, L, MF, SF, Y, PY = load_small_dataset(dataset_name, 'C');
```

```julia
using Printf
```

```julia
view_result(PY, Y, data_state);
```

```julia
view_importance(R, L, MF, SF, show_number=30);
```

```julia
NL = Vector{String}()
for l in L
    if l ∈ [65, 72, 203]
        push!(NL, "*" * string(l))
    else
        push!(NL, string(l))
    end
end
view_importance(R, NL, MF, SF, show_number=30)
d = Dict(zip(L, MF));
@printf "65 : %.4f\n72 : %.4f\n203 : %.4f\n" d[65] d[72] d[203]
```
