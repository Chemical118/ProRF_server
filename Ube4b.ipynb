```julia
include("load_save.jl")
dataset_name = String(split(split(@__FILE__, '\\')[end], '.')[1])
R, X, Y, L, F, Y, PY = load_big_dataset(dataset_name);
```

```julia
using Pandas, Seaborn
```

```julia
view_result(PY, Y, data_state);
```

```julia
view_importance(R, L, F, show_number=30);
```

```julia
NL = Vector{String}()
for l in L
    if l  [1124, 1097, 1107, 1122, 1139, 1142]
        push!(NL, "*" * string(l))
    else
        push!(NL, string(l))
    end
end
view_importance(R, NL, F, show_number=30);
```

```julia
tag = Vector{String}()
for l in L
    if l  [1124, 1097, 1107, 1122, 1139, 1142]
        push!(tag, "High importance region")
    else
        push!(tag, "Remains")
    end
end
data = Pandas.DataFrame(Dict("feature importance" => F, "AA type" => tag));
```

```julia
data["all"] = ""
set_theme(style="whitegrid", palette="tab10")
violinplot(x="all", y="feature importance", hue="AA type", inner="quartile", data=data, split=true, bw=0.4)
xlabel("")
display(gcf())
close("all")
```
